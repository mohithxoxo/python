```
TCS['Volume'].plot(figsize=(16,8),label = "TCS",title='Volume')
INFY['Volume'].plot(label = "INFY")
HCLTECH['Volume'].plot(label = "HCLTECH")
WIPRO['Volume'].plot(label = "WIPRO")
TECHM['Volume'].plot(label = "TECHM")
plt.legend();
```

```
TCS['Total Traded'] = TCS['Open'] * TCS['Volume']
INFY['Total Traded'] = INFY['Open'] * INFY['Volume']
HCLTECH['Total Traded'] = HCLTECH['Open'] * HCLTECH['Volume']
WIPRO['Total Traded'] = WIPRO['Open'] * WIPRO['Volume']
TECHM['Total Traded'] = TECHM['Open'] * TECHM['Volume']
```

```
TCS['Total Traded'].plot(figsize=(16,8),label = "TCS",title='Total Money Traded')
INFY['Total Traded'].plot(label = "INFY")
HCLTECH['Total Traded'].plot(label = "HCLTECH")
WIPRO['Total Traded'].plot(label = "WIPRO")
TECHM['Total Traded'].plot(label = "TECHM")
plt.legend()
plt.ylabel('Total Money Traded')
```

```
TCS['returns'] = TCS['Close'].pct_change(1)
INFY['returns'] = INFY['Close'].pct_change(1)
HCLTECH['returns'] = HCLTECH['Close'].pct_change(1)
WIPRO['returns'] = WIPRO['Close'].pct_change(1)
TECHM['returns'] = TECHM['Close'].pct_change(1)
```

```
TCS['daily_cumulative_return'] = ( 1 + TCS['returns'] ).cumprod()
INFY['daily_cumulative_return'] = ( 1 + INFY['returns'] ).cumprod()
HCLTECH['daily_cumulative_return'] = ( 1 + HCLTECH['returns'] ).cumprod()
WIPRO['daily_cumulative_return'] = ( 1 + WIPRO['returns'] ).cumprod()
TECHM['daily_cumulative_return'] = ( 1 + TECHM['returns'] ).cumprod()
```

```
TCS['daily_cumulative_return'].plot(label='TCS',figsize = (16,8),title = 'Cumilative Return')
INFY['daily_cumulative_return'].plot(label='INFY')
HCLTECH['daily_cumulative_return'].plot(label='HCLTECH')
WIPRO['daily_cumulative_return'].plot(label='WIPRO')
TECHM['daily_cumulative_return'].plot(label='TECHM')
plt.legend();
```
